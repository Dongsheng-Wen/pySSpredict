```python
from sspredict.make_prediction.single_calc_wrapper import single_calc_strength

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import rc, font_manager
ticks_font = font_manager.FontProperties(family='serif', style='normal',
                                         size=24, weight='normal', stretch='normal')
```

```python
# Initialization 
single_calc = single_calc_strength()
```

```python
# prepare a dictonary of data
dict_NiCoFeCr = {"elements":{
                    "Co": {"structure":"fcc","Vn":11.12,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Ni": {"structure":"fcc","Vn":10.94,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Cr": {"structure":"fcc","Vn":12.27,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Fe": {"structure":"fcc","Vn":12.09,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Mn": {"structure":"fcc","Vn":12.60,"G":80,"nu":0.26,"b":2.544,"source":"Literature"}
                    }
                }
```

```python
print('Example 1')
# specify model to use
single_calc.use_model('FCC_Varvenne-Curtin-2016')

# 3d TM CrFeCoNiMn
comp_example1 = {'Co':20,'Fe':20,'Ni':20,'Cr':20,'Mn':20}
# setup the system
single_calc.set_comp(comp_example1)
# specify the adjustable parameters for the model 
single_calc.set_adjustables(f1=0.35, 
                            f2=5.7, 
                            alpha=0.123)

# if not specified, adjustables have default values: 
# single_calc.set_adjustables()

# specify experimental conditions
temperature1 = np.linspace(293,673,20)
strain_r = 0.001
single_calc.set_temp(temperature1)
single_calc.set_strain_r(strain_r)
single_calc.exp_conditions()

# get data from user input
single_calc.grab_data(from_dict=dict_NiCoFeCr)
#print(single_calc.data_of_ssmodels)

# next make prediction
single_calc.calculate()
# if don't want to use model, remove it by
single_calc.remove_model('FCC_Varvenne-Curtin-2016')
single_calc.reorganize_data()
single_calc.pretty_calc_data
```

```python
paper_data_CrFeCoNiMn = [122.,79.,50.]
paper_T_CrFeCoNiMn = [293,473,673]

plt.draw()
f = plt.figure(figsize=(12,9))

(ax1) = f.subplots(1,1)
[i.set_linewidth(1.5) for i in ax1.spines.values()]

ax1.set_xlim(200.0,700)
ax1.set_ylim(0,160)
labels=[570,770,970,1170]

colors=np.array([[0,0,0],[30,136,229],[225,193,7],[212,27,93]])/256


mask = single_calc.calc_data_all['model']=='FCC_Varvenne-Curtin-2016'
ax1.plot(single_calc.calc_data_all[mask]['T'],single_calc.calc_data_all[mask]['tau_y_hT']*3.067,'y-',label='high-T multiscale bow out')
ax1.plot(single_calc.calc_data_all[mask]['T'],single_calc.calc_data_all[mask]['tau_y_lT']*3.067,'b-',label='low-T single scale bow out')


ax1.plot(single_calc.calc_data_all[mask]['T'],single_calc.calc_data_all[mask]['tau_y']*3.067,
              'r-',
              label='predicted')
ax1.plot(paper_T_CrFeCoNiMn,paper_data_CrFeCoNiMn,'rs',markersize=12,label='reference')

ax1.tick_params(direction='in',axis='both',length=10,right=True)
for labelx1 in (ax1.get_xticklabels()):
    labelx1.set_fontproperties(ticks_font)
    
for labely1 in (ax1.get_yticklabels()):
    labely1.set_fontproperties(ticks_font)
    
ax1.set_xlabel('T (K)',fontproperties=ticks_font)
ax1.legend(fontsize=24,frameon=False)
ax1.set_ylabel(r'$\Delta \sigma_{ss}$ (MPa)',fontproperties=ticks_font)


```
