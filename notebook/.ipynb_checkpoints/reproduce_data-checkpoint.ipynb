```python
from sspredict.make_prediction.single_calc_wrapper import single_calc_strength

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from matplotlib import rc, font_manager
ticks_font = font_manager.FontProperties(family='serif', style='normal',
                                         size=24, weight='normal', stretch='normal')
```

```python
# initiation
single_calc = single_calc_strength()
```

```python
# available models 
single_calc.ssmodels_all
```

```python
# prepare a dictonary of data
dict_NiCoFeCr = {"elements":{
                    "Co": {"structure":"fcc","Vn":11.12,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Ni": {"structure":"fcc","Vn":10.94,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Cr": {"structure":"fcc","Vn":12.27,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Fe": {"structure":"fcc","Vn":12.09,"G":80,"nu":0.26,"b":2.544,"source":"Literature"},
                    "Mn": {"structure":"fcc","Vn":12.60,"G":80,"nu":0.26,"b":2.544,"source":"Literature"}
                    }
                }

```

```python
print('Example 1')
# specify model to use
single_calc.use_model('FCC_Varvenne-Curtin-2016')

# 3d TM CrFeCoNiMn
comp_example1 = {'Co':20,'Fe':20,'Ni':20,'Cr':20,'Mn':20}
# setup the system
single_calc.set_comp(comp_example1)
# specify the adjustable parameters for the model 
single_calc.set_adjustables(f1=0.35, 
                            f2=5.7, 
                            alpha=0.123)

# if not specified, adjustables have default values: 
# single_calc.set_adjustables()

# specify experimental conditions
temperature1 = np.linspace(293,673,20)
strain_r = 0.001
single_calc.set_temp(temperature1)
single_calc.set_strain_r(strain_r)
single_calc.exp_conditions()

# get data from user input
single_calc.grab_data(from_dict=dict_NiCoFeCr)
#print(single_calc.data_of_ssmodels)

# next make prediction
single_calc.calculate()
# if don't want to use model, remove it by
single_calc.remove_model('FCC_Varvenne-Curtin-2016')
single_calc.reorganize_data()
single_calc.pretty_calc_data
```

```python
paper_data_CrFeCoNiMn = [122.,79.,50.]
paper_T_CrFeCoNiMn = [293,473,673]

mask = single_calc.calc_data_all['model']=='FCC_Varvenne-Curtin-2016'
plt.plot(single_calc.calc_data_all[mask]['T'],single_calc.calc_data_all[mask]['tau_y']*3.067,'r-',label='predicted tau_y')
plt.plot(paper_T_CrFeCoNiMn,paper_data_CrFeCoNiMn,'ro',label='paper data')
plt.ylim(0,150)
plt.xlim(250,700)
plt.legend()
```

```python
# now to calculate the 1273 K jog mode
T_liquidus_TiNbZr = 1961.
```

```python
single_calc.use_model('BCC_screw_Suzuki_RWASM-2020')

# setup another system: TiNbZr
comp_example4 = {'Nb':32,'Ti':33,'Zr':34,'IT':0.54}
single_calc.set_comp(comp_example4)
single_calc.grab_data(from_dict=dict_TiNbZr)
# specify the adjustable parameters for the model 
single_calc.set_adjustables(kink_width=10, 
                             tau_i_exponent = 1, 
                             dislocation_density= 4e13, 
                             trial_kappa={'min': 1, 'max': 4, 'inc': 0.5}, 
                             trial_tau_k = 5)
# set liquidus temperature
single_calc.set_T_l(T_liquidus_TiNbZr)

single_calc.set_temp(1273)
# set strain rate 
# paper used 0.001 and 0.0001 for 1273 K 
# so for the resulting tau_y, take the average.
for strain_rate in [0.001,0.0001]:
    single_calc.set_strain_r(strain_rate)
    single_calc.calculate()



single_calc.remove_model('BCC_screw_Suzuki_RWASM-2020')

single_calc.reorganize_data()
mask = single_calc.pretty_calc_data['model']=='BCC_screw_Suzuki_RWASM-2020'
single_calc.pretty_calc_data[mask]
```

```python
paper_data_TiNbZr = pd.read_csv('./TiINbZr_paper_extracted.csv',sep='\s+')


plt.plot(single_calc.calc_data_all[mask]['T'],single_calc.calc_data_all[mask]['tau_y']*2.76,'bs-',label='predicted tau_y')
plt.plot(paper_data_TiNbZr['x'],paper_data_TiNbZr['y'],'ro',label='paper data')
plt.ylim(0,1200)
#plt.xlim(250,700)
plt.ylabel(r'$\tau_y$')
plt.legend()
```
